```python
#AND게이트
import numpy as np
x = np.array([0,1]) #입력
w = np.array([0.5,0.5]) #가중치
b = -0.7 #편향(세타)
print(np.sum(w*x)+b)

#y = 0 : b+w1x1 +w2x2 =< 0
#y = 1 : b+w1x1 +w2x2 > 0
```

```python
#함수로 구현(AND 게이트)
def AND(x1, x2):
  x = np.array([x1,x2]) #입력
  w = np.array([0.5,0.5]) #가중치
  b = -0.7 #편향(세타)
  tmp = np.sum(w*x) +b
  if tmp <= 0:
    return 0
  else:
    return 1
```

```python
#함수로 구현(NAND 게이트)
def NAND(x1, x2):
  x = np.array([x1,x2]) #입력
  w = np.array([0.5,0.5]) #가중치
  b = 0.7 #편향(세타)
  tmp = np.sum(w*x) +b
  if tmp <= 0:
    return 0
  else:
    return 1
```

```python
#함수로 구현(OR 게이트)
def OR(x1, x2):
  x = np.array([x1,x2]) #입력
  w = np.array([0.5,0.5]) #가중치
  b = -0.2 #편향(세타)
  tmp = np.sum(w*x) +b
  if tmp <= 0:
    return 0
  else:
    return 1
```

```python
#XOR 게이트 (다중퍼셉트론)
def XOR(x1, x2):
  s1 = NAND(x1,x2)
  s2 = OR(x1, x2)
  y = AND(s1,s2)
  return y

#한번에 안되기 때문에 NAND로 한번(s1), OR로 한번(s2) 그 두값을 AND로 연결하면 됨
```
